```python
import pyodbc
import pandas as pd
import sqlite3

# 1. Connect to SQL Server
sql_conn = pyodbc.connect('DRIVER={SQL Server};SERVER=your_server;DATABASE=your_db;Trusted_Connection=yes;')

# 2. Read the entire table into a Dataframe
df = pd.read_sql_query("SELECT * FROM YourTable", sql_conn)

# 3. Write it directly to SQLite
sqlite_conn = sqlite3.connect('local_database.db')
df.to_sql('TargetTable', sqlite_conn, if_exists='replace', index=False)

# 4. Clean up
sqlite_conn.close()
sql_conn.close()
```
